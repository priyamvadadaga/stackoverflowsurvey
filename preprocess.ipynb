```python
import pandas as pd

survey = pd.read_csv('data/survey_results_public.csv')
```

```python
AI_cols = ['AIToolCurrently Using', 'AIToolInterested in Using', 'AIToolNot interested in Using']
survey_AI = survey[AI_cols]
survey_AI
```

```python
options = [
    "Learning about a codebase", "Project planning", "Writing code",
    "Documenting code", "Debugging and getting help", "Testing code",
    "Committing and reviewing code", "Deployment and monitoring",
    "Predictive analytics", "Search for answers", "Generating content or synthetic data"
]

counts = {opt: {"Currently Using": 0, "Interested in Using": 0, "Not interested in Using": 0} for opt in options}
for _, row in survey_AI.iterrows():
    for col, label in zip(["AIToolCurrently Using", "AIToolInterested in Using", "AIToolNot interested in Using"],
                          ["Currently Using", "Interested in Using", "Not interested in Using"]):
        if pd.notnull(row[col]):
            selected_options = row[col].split(";")
            for opt in selected_options:
                opt = opt.strip()
                if opt in counts:
                    counts[opt][label] += 1

result_df = pd.DataFrame.from_dict(counts, orient='index').reset_index()
```

```python
result_df.to_csv('data/parsed_AItools.csv')
```
