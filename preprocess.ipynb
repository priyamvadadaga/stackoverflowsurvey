```python
import pandas as pd

survey = pd.read_csv('data/survey_results_public.csv')
```

```python
for col in survey.columns:
    print(col)
```

```python
AI_cols = ['AIToolCurrently Using', 'AIToolInterested in Using', 'AIToolNot interested in Using']
survey_AI = survey[AI_cols]
survey_AI
```

```python
options = [
    "Learning about a codebase", "Project planning", "Writing code",
    "Documenting code", "Debugging and getting help", "Testing code",
    "Committing and reviewing code", "Deployment and monitoring",
    "Predictive analytics", "Search for answers", "Generating content or synthetic data"
]

counts = {opt: {"Currently Using": 0, "Interested in Using": 0, "Not interested in Using": 0} for opt in options}
for _, row in survey_AI.iterrows():
    for col, label in zip(["AIToolCurrently Using", "AIToolInterested in Using", "AIToolNot interested in Using"],
                          ["Currently Using", "Interested in Using", "Not interested in Using"]):
        if pd.notnull(row[col]):
            selected_options = row[col].split(";")
            for opt in selected_options:
                opt = opt.strip()
                if opt in counts:
                    counts[opt][label] += 1

result_df = pd.DataFrame.from_dict(counts, orient='index').reset_index()
```

```python
result_df.to_csv('data/parsed_AItools.csv')
```

```python
import pycountry

# Dictionary to store already found country codes
country_code_cache = {}

# Function to get the ISO 3166-1 alpha-3 country code with enhanced fuzzy matching
def get_country_code(country_name):
    # Check if the country code is already cached
    if country_name in country_code_cache:
        return country_code_cache[country_name]
    
    try:
        # First try to get the exact country match
        country = pycountry.countries.get(name=country_name)
        if country:
            country_code_cache[country_name] = country.alpha_3  # Cache the result
            return country.alpha_3
        else:
            # If not found, try fuzzy matching
            possible_matches = pycountry.countries.search_fuzzy(country_name)
            if possible_matches:
                country_code_cache[country_name] = possible_matches[0].alpha_3  # Cache the result
                return possible_matches[0].alpha_3
            else:
                # If no match found, cache None
                country_code_cache[country_name] = None
                return None
    except Exception as e:
        print(f"Error processing {country_name}: {e}")
        country_code_cache[country_name] = None  # Cache None in case of error
        return None

# Apply the function to the 'Country' column to create 'Country_Code' column
survey['Country_Code'] = survey['Country'].apply(get_country_code)

survey[['Country_Code', 'AISelect']].to_csv('AISelect_CountryCode.csv')
```
